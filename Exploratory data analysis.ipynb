## Load and Explore Dataset
In this section, the code loads a dataset from a CSV file and performs exploratory data analysis (EDA) to understand the data better. It displays the first few rows, summary statistics, and checks for missing values in the dataset. Visualizations such as countplot, histograms, boxplots, and correlation heatmap are generated to explore the distribution of data, relationships between features, and their impact on the target variable "Attrition".

```python
# Load the dataset
url = r"C:\Users\Sonymaths\Downloads\archive\WA_Fn-UseC_-HR-Employee-Attrition.csv"
df = pd.read_csv(url)

# Display the first few rows
print(df.head())

# Summary statistics
print(df.describe())

# Check for missing values
print(df.isnull().sum())

# Distribution of target variable (Attrition)
sns.countplot(data=df, x='Attrition')
plt.title('Distribution of Attrition')
plt.show()

# Distribution of numerical features
df.hist(bins=20, figsize=(20, 15))
plt.show()

# Select only numeric columns
numeric_columns = df.select_dtypes(include=[np.number]).columns

# Determine the number of rows and columns for the subplots
num_columns = len(numeric_columns)
num_rows = (num_columns // 5) + (1 if num_columns % 5 != 0 else 0)

# Plot boxplots for each numeric column against 'Attrition'
plt.figure(figsize=(20, 15))
for i, column in enumerate(numeric_columns):
    plt.subplot(num_rows, 5, i + 1)
    sns.boxplot(x='Attrition', y=column, data=df)
    plt.title(f'{column} vs. Attrition')

plt.tight_layout()
plt.show()
plt.figure(figsize=(20, 15))
heatmap = sns.heatmap(
    numeric_df.corr(), 
    annot=True, 
    cmap='viridis',  # Change color palette here
    fmt='.2f', 
    linewidths=.5, 
    linecolor='black',
    cbar_kws={"shrink": .8},
    annot_kws={"size": 10, "weight": "bold"}
)
plt.title('Correlation Matrix', fontsize=20)
plt.xticks(fontsize=12, rotation=45, ha='right')
plt.yticks(fontsize=12)
plt.show()
```
